#### I dropped columns that are empty. The columns include 'bought_together', 'subtitle', 'price', 'store' and 'author'

```python
df_tidy = df.drop(columns = {'bought_together','subtitle','author', 'price', 'store'}) #this columns dropped have missed data and they are not useful in my further analysis
df_tidy
```

#### Display rows with missing values in main_category, average_rating, or rating_number

```python
missing_rows = df_tidy[df_tidy[['main_category', 'average_rating', 'rating_number']].isnull().any(axis=1)]
print(missing_rows) 
```

```python
# Drop rows where 'average_rating' is missing
df_tidy = df_tidy.dropna(subset=['average_rating'])
```

```python
df_tidy
```

```python
# Checking to see if there is any missing values in 'average_rating'
print(df_tidy['average_rating'].isnull().sum())
```

```python
df_tidy = df_tidy.copy()
```

```python
df_tidy['rating_number'] = df_tidy['rating_number'].fillna(0)

df_tidy
```

```python
# Checking to see if there is any missing values in 'rating_number' 
print(df_tidy['rating_number'].isnull().sum())
```

```python
# Replacing missing values in the 'main_category' column with unknown as it is important for my further analysis and also prevent loss of potential information
df_tidy['main_category'] = df_tidy['main_category'].fillna('Unknown')

df_tidy
```

```python
# Checking to see if there is any missing values in 'rating_number'
print(df_tidy['main_category'].isnull().sum())
```
